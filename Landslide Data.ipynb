<b>Replace headers (Data Wrangling with Python pg. 154 –163)</b><br/>
The headers have trailing '_', a mix of uppercase and lowercase, and are not capitalized.

<b>Conduct Fuzzy Matching (if you don’t have an obvious example to do this with in your data, create categories and use Fuzzy Matching to lump data together) (Data Wrangling with Python pg. 179 –188)</b>

In this step, I want to merge the capitalized and non-capitalized versions of each of my values (i.e. 'small' and 'Small') so that they appear as the same value. There were also many entries listed as 'Large' or 'Very_large' and just one entry listed as 'Extra large'. Since we don't know the distinction between them, I want to simplify the scale down to three options:
<ul>
<li>Small</li>
<li>Medium</li>
<li>Large</li>
</ul>

Since the values are clearly distinct, fuzzy-matching is not really required here, but I am going to use it to show how certain I am of which values belong together before I merge them.

```python
# Open .txt file containing all valid sizes
with open("Landslide_Sizes.txt", "r") as f:
    sizes = f.read().split("\n")

# use fuszzywuzzy
from fuzzywuzzy import process

def get_matches(query, choices, limit=2):
    results = process.extract(query, choices, limit = limit)
    return results

print("Check for 'small': " + str(get_matches("small", sizes)))
print(">>>100% match with 'Small'")

print("\nCheck for 'medium': " + str(get_matches("medium", sizes)))
print(">>>100% match with 'Medium'")

print("\nCheck for 'large': " + str(get_matches("large", sizes)))
print(">>>100% match with 'Large'")

print("\nCheck for 'Very_large': " + str(get_matches("Very_large", sizes)))
print(">>>90% match with 'Large'")

print("\nCheck for 'Extra large': " + str(get_matches("Extra large", sizes)))
print(">>>90% match with 'Large'")

# Rename values
df['Landslide_Size'].replace(['small'],'Small', inplace=True)
df['Landslide_Size'].replace(['medium'],'Medium', inplace=True)
df['Landslide_Size'].replace(['large'],'Large', inplace=True)
df['Landslide_Size'].replace(['Very_large'],'Large', inplace=True)
df['Landslide_Size'].replace(['Extra Large'],'Large', inplace=True)

print("\n\nFINAL VALUE CATEGORIES: " + str(df['Landslide_Size'].unique()))
```

<b>Recall our question for analysis:</b> In the past few years, what continent has experienced the biggest landslides the most frequently?

```python
# View list of all continents in the data
# print("Continents: " + str(df['Continentc'].unique()))

# Create variables to store data from individual continents
df_NA = df[(df['Continentc']=="NA") & (df['Landslide_Size']=="Large")]
df_AS = df[(df['Continentc']=="AS") & (df['Landslide_Size']=="Large")]
df_SA = df[(df['Continentc']=="SA") & (df['Landslide_Size']=="Large")]
df_OC = df[(df['Continentc']=="OC") & (df['Landslide_Size']=="Large")]
df_EU = df[(df['Continentc']=="EU") & (df['Landslide_Size']=="Large")]
df_AF = df[(df['Continentc']=="AF") & (df['Landslide_Size']=="Large")]

# Create list of continents
continentList = df['Continentc'].unique()

# Create list to store quantities
quantityList = [len(df_NA.index), len(df_AS.index), len(df_SA.index), len(df_OC.index), len(df_EU.index), len(df_AF.index)]

max_value = max(quantityList)
max_index = quantityList.index(max_value)

print("Continent with the most frequent 'biggest' landslides: " + str(continentList[max_index]))
print("Number of 'biggest' landslides in that continent: " + str(max_value))
```
